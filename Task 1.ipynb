```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_df = sns.load_dataset('iris')
```

```python
# Display concise summary of the DataFrame
print("\nDataFrame Info:")
iris_df.info()
```

```python
# Display descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(iris_df.describe())
```

```python
# Create a scatter plot to show relationships between features, colored by species
fig = sns.pairplot(iris_df, hue='species', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot of Iris Dataset Features by Species', y=1.02) # Add a title to the figure
plt.show()
```

```python
# Create histograms for each numerical feature
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Histograms of Iris Dataset Features', y=1.02)

sns.histplot(iris_df['sepal_length'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Sepal Length Distribution')

sns.histplot(iris_df['sepal_width'], kde=True, ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title('Sepal Width Distribution')

sns.histplot(iris_df['petal_length'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Petal Length Distribution')

sns.histplot(iris_df['petal_width'], kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Petal Width Distribution')

plt.tight_layout()
plt.show()
```

```python
# Create box plots for each numerical feature to identify outliers
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Box Plots of Iris Dataset Features by Species', y=1.02)

sns.boxplot(x='species', y='sepal_length', data=iris_df, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Sepal Length by Species')

sns.boxplot(x='species', y='sepal_width', data=iris_df, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Sepal Width by Species')

sns.boxplot(x='species', y='petal_length', data=iris_df, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Petal Length by Species')

sns.boxplot(x='species', y='petal_width', data=iris_df, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```
